<h2>Monthly Sales per Year</h2>

```python
import calendar

df['MONTH_NAME'] = df['MONTH_ID'].apply(lambda x: calendar.month_abbr[int(x)])
```

```python
month_order = list(calendar.month_abbr)[1:]   # ['Jan','Feb',...,'Dec']

df['MONTH_NAME'] = pd.Categorical(df['MONTH_NAME'], 
                                  categories=month_order, 
                                  ordered=True)
```

```python
pivot = df.groupby(['YEAR_ID', 'MONTH_NAME'])['SALES'].sum().reset_index()

pivot_table = pivot.pivot(index='MONTH_NAME', columns='YEAR_ID', values='SALES')
pivot_table = pivot_table.replace(0, np.nan)

pivot_table
```

```python
pivot_table.plot(figsize=(10,5) , marker = 'o')

plt.title("Monthly Sales per Year")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.legend(title="Year")
plt.show()
```

<h2>RFM</h2>

```python
snapshot = df['ORDERDATE'].max() + pd.Timedelta(days=1)

rfm = df.groupby('CUSTOMERNAME').agg({
    'ORDERDATE': lambda x: (snapshot - x.max()).days,
    'ORDERNUMBER': 'nunique',
    'SALES': 'sum'
})

rfm.columns = ['Recency','Frequency','Monetary']
rfm
```

```python
rfm['R_Score'] = pd.qcut(rfm['Recency'].rank(method='first'), 
                         5, labels=[5,4,3,2,1])
rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method='first', ascending=False), 
                         5, labels=[1,2,3,4,5])

rfm['M_Score'] = pd.qcut(rfm['Monetary'].rank(method='first', ascending=False), 
                         5, labels=[1,2,3,4,5])

rfm['RFM_Score'] = (
    rfm['R_Score'].astype(str) + 
    rfm['F_Score'].astype(str) + 
    rfm['M_Score'].astype(str)
)
rfm.head()
```

```python
def segment(row):
    r = int(row['R_Score'])
    f = int(row['F_Score'])
    m = int(row['M_Score'])

    if r >= 4 and f >= 4 and m >= 4:
        return 'VIP'
    if r >= 3 and f >= 3:
        return 'Potential Loyalist'
    if r <= 2 and f <= 2:
        return 'At Risk'
    return 'Needs Attention'

rfm['Segment'] = rfm.apply(segment, axis=1)
rfm.head()
```

<h3>Overview Cards</h3>
